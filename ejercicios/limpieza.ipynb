```python
import pandas as pd
import os

ruta=r"C:\Users\DAVID01\Documents\GitHub\Geospatial-analysis\ejercicios\carto\estaciones.csv"
ruta_series=r"C:\Users\DAVID01\Documents\GitHub\Geospatial-analysis\data\series limpias"
df=pd.read_csv(ruta,sep=';')

codigos=df['Codigo'].to_list()

archivos=os.listdir(ruta_series)

for file in archivos:
    if file.endswith('.csv'):
        try:
            cod=int(os.path.splitext(file)[0].split('_')[0])
            if cod in codigos:
                continue
            else:
                os.remove(os.path.join(ruta_series,file))
        except Exception as e:
            print(f'Error procesando {file}: {e}')
```
